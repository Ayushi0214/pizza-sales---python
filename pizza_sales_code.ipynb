```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv("orders.csv")
orders
```

```python
o_details = pd.read_csv("order_details.csv")
o_details
```

```python
p_types = pd.read_csv("pizza_types.csv",encoding='unicode_escape')
p_types
```

```python
pizzas = pd.read_csv("pizzas.csv")
pizzas
```

```python
df = pd.merge(left = o_details, right = pizzas, on = "pizza_id", how = "left")
df
```

```python
df1 = pd.merge(left = df, right = p_types, on = "pizza_type_id", how = "left")
df1
```

```python
orders.info()
```

```python
orders["date"] = pd.to_datetime(orders["date"])
orders.info()
```

```python
orders["time"] = pd.to_datetime(orders["time"])
orders.info()
```

# total sales

```python
df1["Amount"] = df1["quantity"] * df1["price"]
df1["Amount"].sum()
```

# most ordered pizza

```python
sns.countplot(x = "name", data = df1)
plt.xticks(rotation = 90)
plt.show()
```

```python
gb = df1.groupby("name").agg({"order_details_id":"count"})
gb = gb.sort_values(by = "order_details_id", ascending = False )
print("the most ordered pizza is", gb["order_details_id"].idxmax())
```

# Hotselling pizza

```python
gb = df1.groupby("name").agg({"Amount":"sum"})
gb = gb.sort_values(by = "Amount", ascending = False )
print("the most ordered pizza is", gb["Amount"].idxmax())
```

# hotselling category

```python
gb = df1.groupby("category").agg({"Amount":"sum"})
gb = gb.sort_values(by = "Amount", ascending = False )
print("the most ordered pizza is", gb["Amount"].idxmax())
```

```python
plt.pie(gb["Amount"], labels = gb.index, autopct = "%1.2f%%")
plt.show()
```

# busiest day of the week

```python
orders["weekday"] = orders["date"].dt.strftime("%A")
orders
```

```python
gb = orders.groupby("weekday").agg({"order_id":"count"})
gb
```

# identify the time at which we are getting high orders and the time at which we are getting the lowest orders

```python
orders["hour"] = orders["time"].dt.strftime("%H")
orders
```

```python
gb = orders.groupby("hour").agg({'order_id':"count"})
gb
```

```python
plt.plot( gb.index, gb["order_id"], marker = "o")
plt.show()
```

```python
#Identify at which time of the day, people are ordering more pizzas.       (morning/evening/afternoon/night)


def a(value):
    if value < 12:
        return "morning"
    elif value >= 12 and value < 16:
        return "afternoon"
    elif value >= 16 and value < 20:
        return "evening"
    else:
        return "night"

orders["day_time"] = orders["hour"].apply(a)
orders
```

```python
orders['hour'] = orders["hour"].astype('int')
```

```python
gb = orders.groupby("day_time").agg({"order_id":"count"})
gb
```
